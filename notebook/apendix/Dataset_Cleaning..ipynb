```python
# ================================
# Binary Target: Merge, Cap Outliers, Save CSV
# ================================
import pandas as pd
import numpy as np
import os

binary_path = "data/binary_target/"
merged_path = "data/merged/"
os.makedirs(merged_path, exist_ok=True)

# --- Merge CSVs ---
csv_files = [f for f in os.listdir(binary_path) if f.lower().endswith('.csv')]
dfs = []

for file in csv_files:
    df = pd.read_csv(os.path.join(binary_path, file))
    dfs.append(df)
    print(f"Loaded {file} with shape {df.shape}")

binary_df = pd.concat(dfs, ignore_index=True)
print(f"✅ Binary dataset merged. Shape: {binary_df.shape}")

# --- Cap Outliers Function ---
def cap_outliers_iqr(df, exclude_cols=None):
    if exclude_cols is None:
        exclude_cols = []
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    numeric_cols = [c for c in numeric_cols if c not in exclude_cols]

    for col in numeric_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        df[col] = df[col].clip(lower, upper)

    print(f"✅ Outlier capping done for {len(numeric_cols)} features.")
    return df

# --- Apply capping ---
binary_df_clean = cap_outliers_iqr(binary_df, exclude_cols=['target_binary', 'Scenario_ID'])

# --- Save Cleaned CSV ---
binary_df_clean.to_csv(os.path.join(merged_path, "binary_dataset_clean.csv"), index=False)
print(f"✅ Binary cleaned dataset saved. Shape: {binary_df_clean.shape}")
```

```python
# ================================
# Three-Class Target: Merge, Cap Outliers, Save CSV
# ================================
three_class_path = "data/three_class_target/"

csv_files = [f for f in os.listdir(three_class_path) if f.lower().endswith('.csv')]
dfs = []

for file in csv_files:
    df = pd.read_csv(os.path.join(three_class_path, file))
    dfs.append(df)
    print(f"Loaded {file} with shape {df.shape}")

three_class_df = pd.concat(dfs, ignore_index=True)
print(f"✅ Three-Class dataset merged. Shape: {three_class_df.shape}")

three_class_df_clean = cap_outliers_iqr(three_class_df, exclude_cols=['target_three_class', 'Scenario_ID'])

three_class_df_clean.to_csv(os.path.join(merged_path, "three_class_dataset_clean.csv"), index=False)
print(f"✅ Three-Class cleaned dataset saved. Shape: {three_class_df_clean.shape}")
```

```python
# ================================
# Multi-Class Target: Merge ARFF, Cap Outliers, Save CSV
# ================================
import arff

multi_class_path = "data/multi_class_target/"

arff_files = [f for f in os.listdir(multi_class_path) if f.lower().endswith('.arff')]
dfs = []

for file in arff_files:
    with open(os.path.join(multi_class_path, file), 'r') as f:
        data = arff.load(f)
        df = pd.DataFrame(data['data'], columns=[attr[0] for attr in data['attributes']])
        dfs.append(df)
        print(f"Loaded {file} with shape {df.shape}")

multi_class_df = pd.concat(dfs, ignore_index=True)
print(f"✅ Multi-Class dataset merged. Shape: {multi_class_df.shape}")

multi_class_df_clean = cap_outliers_iqr(multi_class_df, exclude_cols=['target_multi_class', 'Scenario_ID'])

multi_class_df_clean.to_csv(os.path.join(merged_path, "multi_class_dataset_clean.csv"), index=False)
print(f"✅ Multi-Class cleaned dataset saved as CSV. Shape: {multi_class_df_clean.shape}")
```

```python
import pandas as pd
import numpy as np

# ============================================================
# 0. LOAD RAW MERGED DATASET (NO CLEANING DONE YET)
# ============================================================

df = pd.read_csv("../data/merged/multi_class_dataset_clean.csv")
print("Loaded raw dataset:", df.shape)

# ============================================================
# 1. DETECT COLUMNS CONTAINING INF
# ============================================================

Z_cols = ["R1-PA:Z", "R2-PA:Z", "R3-PA:Z", "R4-PA:Z"]

print("\nChecking inf values:")
for col in Z_cols:
    print(f"{col}: {np.isinf(df[col]).sum()} inf values")

# ============================================================
# 2. ADD Z_infinity FLAGS  (PHYSICALLY IMPORTANT)
# ============================================================

# These flags preserve the info that impedance blew up (breaker open, relay disabled, etc.)
for col in Z_cols:
    df[col + "_inf_flag"] = np.isinf(df[col]).astype(int)

print("\nAdded Z_inf_flag columns.")

# ============================================================
# 3. TEMPORARILY REPLACE INF WITH NaN (SO WE CAN CAP PROPERLY)
# ============================================================

df = df.replace([np.inf, -np.inf], np.nan)

# ============================================================
# 4. CAP EXTREME IMPEDANCE VALUES (ENGINEERING-CORRECT)
# ------------------------------------------------------------
# NOTE: This is SAFE because we are NOT computing the cap now.
# We cap LATER per TRAIN SPLIT to avoid leakage.
# Right now we simply mark NaN (former inf values).
# ============================================================

# DO NOT CAP NOW.  Save NaN values so train/val/test split can handle them.

# ============================================================
# 5. REMOVE ROWS CONTAINING ANY NaN EXCEPT Z COLUMNS
# ------------------------------------------------------------
# We keep NaN in Z columns because we haven't capped yet.
# We only remove NaN in PMU voltage/current features.
# ============================================================

# Identify allowed NaN columns (Z)
allowed_nan_cols = set(Z_cols)

cols_with_nan = df.columns[df.isna().any()].tolist()

cols_to_drop_na_from = [c for c in cols_with_nan if c not in allowed_nan_cols]

df_clean = df.dropna(subset=cols_to_drop_na_from)

print("\nRows removed due to NaN in non-Z columns:", len(df) - len(df_clean))
print("Dataset after first cleaning:", df_clean.shape)

# ============================================================
# 6. SAVE TEMP CLEANED DATASET — STILL NEEDS CAPPING LATER
# (This is NOT final; capping must be applied AFTER train split)
# ============================================================

df_clean.to_csv("../data/merged/multi_class_dataset_clean_intermediate.csv", index=False)
print("\nSaved intermediate cleaned dataset.")
```

```python
import pandas as pd
import numpy as np

INPUT = "../data/merged/multi_class_dataset_clean.csv"
OUTPUT = "../data/merged/multi_class_dataset_clean_FULL.csv"

df = pd.read_csv(INPUT)
print("Loaded raw dataset:", df.shape)

# 1. Detect INF values
z_columns = ["R1-PA:Z", "R2-PA:Z", "R3-PA:Z", "R4-PA:Z"]

print("\nChecking INF values:")
for col in z_columns:
    inf_count = np.isinf(df[col]).sum()
    print(f"{col}: {inf_count} inf values")
    df[col + "_inf_flag"] = np.isinf(df[col]).astype(int)

# 2. Replace INF with NaN
df = df.replace([np.inf, -np.inf], np.nan)

# 3. Fill NaN with median
df = df.fillna(df.median(numeric_only=True))

print("\nAfter full cleaning:", df.shape)
print("INF left:", np.isinf(df.values).sum())
print("NaN left:", df.isna().sum().sum())

# Save final cleaned dataset
df.to_csv(OUTPUT, index=False)
print("Saved full cleaned dataset to:", OUTPUT)
```

```python
import pandas as pd
import os

OUTPUT_DIR = "../data/datasets_hierarchical/cleaned_final"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Load fully cleaned dataset
df = pd.read_csv("../data/merged/multi_class_dataset_clean_FULL.csv")
print("Loaded FULL cleaned dataset:", df.shape)

# ------------- Scenario groups ----------------
faults = list(range(1, 7))
maintenance = [13, 14]
normal = [41]

data_injection = list(range(7, 13))
remote_tripping = list(range(15, 21))
relay_setting = list(range(21, 31)) + list(range(35, 41))

all_non_attacks = faults + maintenance + normal
all_attacks = data_injection + remote_tripping + relay_setting

# ============================================================
# M1 — ATTACK vs NON-ATTACK (NO MARKER)
# ============================================================
df_m1 = df.copy()
df_m1["label"] = df_m1["marker"].apply(lambda m: 1 if m in all_attacks else 0)
df_m1["label_name"] = df_m1["label"].map({0: "Non-Attack", 1: "Attack"})

df_m1 = df_m1.drop(columns=["marker"])
df_m1.to_csv(f"{OUTPUT_DIR}/M1_Attack_vs_NonAttack_clean.csv", index=False)
print("Saved M1")

# ============================================================
# M2 — NATURAL 3-CLASS (NO MARKER)
# ============================================================
df_m2 = df[df["marker"].isin(all_non_attacks)].copy()

def map_nat(m):
    if m in faults:
        return 0
    elif m in maintenance:
        return 1
    else:
        return 2

df_m2["label"] = df_m2["marker"].apply(map_nat)
df_m2["label_name"] = df_m2["label"].map({
    0: "Short-Circuit Fault",
    1: "Line Maintenance",
    2: "Normal Operation"
})

df_m2 = df_m2.drop(columns=["marker"])
df_m2.to_csv(f"{OUTPUT_DIR}/M2_Natural_3Class_clean.csv", index=False)
print("Saved M2")

# ============================================================
# M3 — ATTACK 3-CLASS (NO MARKER)
# ============================================================
df_m3 = df[df["marker"].isin(all_attacks)].copy()

def map_attack(m):
    if m in data_injection:
        return 0
    elif m in remote_tripping:
        return 1
    else:
        return 2

df_m3["label"] = df_m3["marker"].apply(map_attack)
df_m3["label_name"] = df_m3["label"].map({
    0: "Data Injection",
    1: "Remote Tripping",
    2: "Relay Setting Change"
})

df_m3 = df_m3.drop(columns=["marker"])
df_m3.to_csv(f"{OUTPUT_DIR}/M3_Attack_3Class_clean.csv", index=False)
print("Saved M3")

# ============================================================
# INTERNAL LEVELS: M4–M6 (KEEP MARKER)
# ============================================================
df_m4 = df[df["marker"].isin(data_injection)].copy()
df_m4["label"] = df_m4["marker"]
df_m4.to_csv(f"{OUTPUT_DIR}/M4_DataInjection_Internal_clean.csv", index=False)
print("Saved M4")

df_m5 = df[df["marker"].isin(remote_tripping)].copy()
df_m5["label"] = df_m5["marker"]
df_m5.to_csv(f"{OUTPUT_DIR}/M5_RemoteTripping_Internal_clean.csv", index=False)
print("Saved M5")

df_m6 = df[df["marker"].isin(relay_setting)].copy()
df_m6["label"] = df_m6["marker"]
df_m6.to_csv(f"{OUTPUT_DIR}/M6_RelaySetting_Internal_clean.csv", index=False)
print("Saved M6")

print("\n🎉 All hierarchical datasets regenerated successfully!")
```
