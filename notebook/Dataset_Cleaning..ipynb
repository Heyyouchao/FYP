```python
# ================================
# Binary Target: Merge, Cap Outliers, Save CSV
# ================================
import pandas as pd
import numpy as np
import os

binary_path = "data/binary_target/"
merged_path = "data/merged/"
os.makedirs(merged_path, exist_ok=True)

# --- Merge CSVs ---
csv_files = [f for f in os.listdir(binary_path) if f.lower().endswith('.csv')]
dfs = []

for file in csv_files:
    df = pd.read_csv(os.path.join(binary_path, file))
    dfs.append(df)
    print(f"Loaded {file} with shape {df.shape}")

binary_df = pd.concat(dfs, ignore_index=True)
print(f"✅ Binary dataset merged. Shape: {binary_df.shape}")

# --- Cap Outliers Function ---
def cap_outliers_iqr(df, exclude_cols=None):
    if exclude_cols is None:
        exclude_cols = []
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    numeric_cols = [c for c in numeric_cols if c not in exclude_cols]

    for col in numeric_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        df[col] = df[col].clip(lower, upper)

    print(f"✅ Outlier capping done for {len(numeric_cols)} features.")
    return df

# --- Apply capping ---
binary_df_clean = cap_outliers_iqr(binary_df, exclude_cols=['target_binary', 'Scenario_ID'])

# --- Save Cleaned CSV ---
binary_df_clean.to_csv(os.path.join(merged_path, "binary_dataset_clean.csv"), index=False)
print(f"✅ Binary cleaned dataset saved. Shape: {binary_df_clean.shape}")
```

```python
# ================================
# Three-Class Target: Merge, Cap Outliers, Save CSV
# ================================
three_class_path = "data/three_class_target/"

csv_files = [f for f in os.listdir(three_class_path) if f.lower().endswith('.csv')]
dfs = []

for file in csv_files:
    df = pd.read_csv(os.path.join(three_class_path, file))
    dfs.append(df)
    print(f"Loaded {file} with shape {df.shape}")

three_class_df = pd.concat(dfs, ignore_index=True)
print(f"✅ Three-Class dataset merged. Shape: {three_class_df.shape}")

three_class_df_clean = cap_outliers_iqr(three_class_df, exclude_cols=['target_three_class', 'Scenario_ID'])

three_class_df_clean.to_csv(os.path.join(merged_path, "three_class_dataset_clean.csv"), index=False)
print(f"✅ Three-Class cleaned dataset saved. Shape: {three_class_df_clean.shape}")
```

```python
# ================================
# Multi-Class Target: Merge ARFF, Cap Outliers, Save CSV
# ================================
import arff

multi_class_path = "data/multi_class_target/"

arff_files = [f for f in os.listdir(multi_class_path) if f.lower().endswith('.arff')]
dfs = []

for file in arff_files:
    with open(os.path.join(multi_class_path, file), 'r') as f:
        data = arff.load(f)
        df = pd.DataFrame(data['data'], columns=[attr[0] for attr in data['attributes']])
        dfs.append(df)
        print(f"Loaded {file} with shape {df.shape}")

multi_class_df = pd.concat(dfs, ignore_index=True)
print(f"✅ Multi-Class dataset merged. Shape: {multi_class_df.shape}")

multi_class_df_clean = cap_outliers_iqr(multi_class_df, exclude_cols=['target_multi_class', 'Scenario_ID'])

multi_class_df_clean.to_csv(os.path.join(merged_path, "multi_class_dataset_clean.csv"), index=False)
print(f"✅ Multi-Class cleaned dataset saved as CSV. Shape: {multi_class_df_clean.shape}")
```
